```python
import pandas as pd

# The raw URL for your specific CSV file
url = 'https://raw.githubusercontent.com/mhmdparichehreh/basketball-scores/main/data/scores.csv'

# Read the CSV
df = pd.read_csv(url)

# Display the first 5 rows
df.head()
```

```python
df
```

```python
### all the matches: q1,q2,q3/3 > q4
df.iloc[:,9:11].sum(axis=1)
hypo1 = (df.iloc[:,3:9].sum(axis=1)/3) > df.iloc[:,9:11].sum(axis=1)
l1 = hypo1.sum()
print(l1, len(hypo1))
```

```python
### Female matches: q1,q2,q3/3 > q4
df_female = df[df['gender'] == 'F']
df_female.iloc[:,9:11].sum(axis=1)
hypo1 = (df_female.iloc[:,3:9].sum(axis=1)/3) > df_female.iloc[:,9:11].sum(axis=1)
l1 = hypo1.sum()
print(l1, len(hypo1))
```
